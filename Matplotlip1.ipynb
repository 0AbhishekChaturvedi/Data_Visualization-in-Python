```python
batsman=pd.read_csv("/content/sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# multiple plotting
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# customizing: Giving Title, and labbeling axises
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# customizing: color
plt.plot(batsman['index'],batsman['V Kohli'],color="green")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')

plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# customizing: linestyle
# changed the line style of kohili..and since you didnt provide for Rohit, it is 
# solid line by default
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
```

```python
# There are multiple line style... you can check the matplotlib documentation
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')
```

```python
# customization : you can also change the width of line style
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')
```

```python
#Customizing: the marker: it will highlight the points on 2D space.
# again there are multiple option... read documentation
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted')
```

```python
# Customizing: you can also manipulate the size of the markers
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D',markersize=6)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted')
```

```python
# customizing : printing legend .. you cant differenciate between the two 
# graphs above. i.e. how can one know which line represents rohit or virat..
# use label and legend

plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D',markersize=7,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted',label="Rohit")
plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
```

```python
# Customizing : loc of legend-> this gives you option to move your legend..
# by default it is best.. it automatically fits where it seems best to it

plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D',markersize=7,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted',label="Rohit")
plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend(loc="upper right")
```

```python
# limiting axes : Generaly you do
price =[48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
```

```python
# since you have an outlier in your data, your other curves starts looking flat..
# so for looking all variation , you can trim your graph
plt.plot(year,price)
plt.ylim(0,100000)
```

```python
# you can also do this for x axis
plt.plot(year,price)
plt.ylim(0,100000)
plt.xlim(2017,2019)
```

```python
# customizing  grid 
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D',markersize=7,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted',label="Rohit")
plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend(loc="upper left")
plt.grid()
```

```python
# plt.show: if you run this piece of code(same as above) in some text editor like pycharm,
# it wont show any plotting unless you call plt.show() explicitly
# since this(colab jupyter notebook) is console based.. it is showing
plt.plot(batsman['index'],batsman['V Kohli'],color="green",linestyle='dashdot',marker='D',markersize=7,label="Virat")
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',marker="+",linestyle='dotted',label="Rohit")
plt.title("Rohit vs Virat in IPL")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```
